```python
import pandas as pd

path = "/content/drive/MyDrive/Parkinson's Dataset for Mlops Project/parkinsons/parkinsons.data"
df1 = pd.read_csv(path)

df1.head()
```

```python
df3 = pd.read_csv("/content/drive/MyDrive/Parkinson's Dataset for Mlops Project/parkinsons.data")

print(df1.shape)
print(df3.shape)

print(df1.equals(df3))
```

```python
print("Shape:", df1.shape)
print("\nColumns:\n", df1.columns)
df1.head()
```

```python
# Class Distribution
df1['status'].value_counts()
```

```python
#Separate Features and Label
X1 = df1.drop(columns=['status', 'name'])
y1 = df1['status']
```

```python
X1.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,10))
sns.heatmap(X1.corr(), cmap='coolwarm')
plt.title("Correlation Heatmap - UCI Dataset")
plt.show()
```

```python
sns.histplot(data=df1, x='MDVP:Jitter(%)', hue='status', kde=True)
plt.title("Jitter Distribution")
plt.show()

sns.histplot(data=df1, x='MDVP:Shimmer', hue='status', kde=True)
plt.title("Shimmer Distribution")
plt.show()

sns.histplot(data=df1, x='HNR', hue='status', kde=True)
plt.title("HNR Distribution")
plt.show()
```

```python
# Feature Importance
from sklearn.ensemble import RandomForestClassifier

model1 = RandomForestClassifier(random_state=42)
model1.fit(X1, y1)

importances1 = pd.Series(model1.feature_importances_, index=X1.columns)
importances1 = importances1.sort_values(ascending=False)

print(importances1.head(10))
```

```python
importances1.head(10).plot(kind='barh')
plt.title("Top 10 Important Features - UCI")
plt.gca().invert_yaxis()
plt.show()
```

### Dataset 2

```python
train = pd.read_csv("/content/drive/MyDrive/Parkinson's Dataset for Mlops Project/parkinson+speech+dataset+with+multiple+types+of+sound+recordings/train_data.txt", header=None)

print("Shape:", train.shape)
train.head()
```

```python
feature_names = [
"Jitter_local","Jitter_abs","Jitter_rap","Jitter_ppq5","Jitter_ddp",
"Shimmer_local","Shimmer_dB","Shimmer_apq3","Shimmer_apq5","Shimmer_apq11","Shimmer_dda",
"AC","NTH","HTN",
"Median_pitch","Mean_pitch","Std_pitch","Min_pitch","Max_pitch",
"Num_pulses","Num_periods","Mean_period","Std_period",
"Frac_unvoiced","Num_voice_breaks","Degree_voice_breaks"
]

columns = ["Subject_ID"] + feature_names + ["UPDRS","Class"]
train.columns = columns

train.head()
```

```python
train["Class"].value_counts()
```

```python
X2 = train[feature_names]
y2 = train["Class"]
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(X2.corr(), cmap='coolwarm')
plt.title("Correlation Heatmap - Istanbul Dataset")
plt.show()
```

```python
sns.histplot(data=train, x='Jitter_local', hue='Class', kde=True)
plt.title("Jitter_local Distribution")
plt.show()

sns.histplot(data=train, x='Shimmer_local', hue='Class', kde=True)
plt.title("Shimmer_local Distribution")
plt.show()

sns.histplot(data=train, x='Mean_pitch', hue='Class', kde=True)
plt.title("Mean Pitch Distribution")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

model2 = RandomForestClassifier(random_state=42)
model2.fit(X2, y2)

importances2 = pd.Series(model2.feature_importances_, index=X2.columns)
importances2 = importances2.sort_values(ascending=False)

print(importances2.head(10))
```

```python
importances2.head(10).plot(kind='barh')
plt.title("Top 10 Important Features - Istanbul")
plt.gca().invert_yaxis()
plt.show()
```

```python
corr_motor = df3.corr()["motor_UPDRS"].sort_values(ascending=False)
corr_total = df3.corr()["total_UPDRS"].sort_values(ascending=False)

print("Top correlations with motor_UPDRS:")
print(corr_motor.head(10))

print("\nTop correlations with total_UPDRS:")
print(corr_total.head(10))
```
